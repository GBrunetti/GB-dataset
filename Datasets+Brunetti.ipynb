```python
import pandas as pd
```

```python
file_path1 = "BankChurners.csv"
```

```python
datos1 = pd.read_csv(file_path1)
```

```python
plt.figure(figsize=(12, 6))
plt.hist(datos1['Customer_Age'], bins=20, color='skyblue', edgecolor='black')
plt.axvline(datos1['Customer_Age'].mean(), color='red', linestyle='dashed', linewidth=1, label='Media de Edad')
plt.xlabel('Edad del Cliente')
plt.ylabel('Frecuencia')
plt.title('DistribuciÃ³n de Edades de los Clientes')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
pd.crosstab(datos1['Marital_Status'], datos1['Attrition_Flag']).plot(kind='bar', stacked=True, color=['skyblue', 'salmon'])
plt.xlabel('Estado Civil')
plt.ylabel('Cantidad de Clientes')
plt.title('DistribuciÃ³n de Clientes por Estado Civil y CategorÃ­a de Abandono')
plt.legend(title='Attrition_Flag', labels=['Existing Customer', 'Attrited Customer'])
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
pd.crosstab(datos1['Card_Category'], datos1['Attrition_Flag']).plot(kind='bar', stacked=True, color=['skyblue', 'salmon'])
plt.xlabel('CategorÃ­a de Tarjeta de CrÃ©dito')
plt.ylabel('Cantidad de Clientes')
plt.title('DistribuciÃ³n de Clientes por CategorÃ­a de Tarjeta de CrÃ©dito y CategorÃ­a de Abandono')
plt.legend(title='Attrition_Flag', labels=['Existing Customer', 'Attrited Customer'])
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(datos1[datos1['Attrition_Flag']=='Existing Customer']['Customer_Age'], bins=20, color='skyblue', alpha=0.7, label='Existing Customer')
plt.hist(datos1[datos1['Attrition_Flag']=='Attrited Customer']['Customer_Age'], bins=20, color='salmon', alpha=0.7, label='Attrited Customer')
plt.xlabel('Edad del Cliente')
plt.ylabel('Frecuencia')
plt.title('DistribuciÃ³n de Edades de Clientes por CategorÃ­a de Abandono')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.boxplot([datos1[datos1['Attrition_Flag']=='Existing Customer']['Credit_Limit'],
             datos1[datos1['Attrition_Flag']=='Attrited Customer']['Credit_Limit']],
            labels=['Existing Customer', 'Attrited Customer'], showmeans=True)
plt.ylabel('LÃ­mite de CrÃ©dito')
plt.title('DistribuciÃ³n de LÃ­mites de CrÃ©dito por CategorÃ­a de Abandono')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
pd.crosstab(datos1['Income_Category'], datos1['Attrition_Flag']).plot(kind='bar', stacked=True, color=['skyblue', 'salmon'])
plt.xlabel('CategorÃ­a de Ingresos')
plt.ylabel('Cantidad de Clientes')
plt.title('DistribuciÃ³n de Clientes por CategorÃ­a de Ingresos y CategorÃ­a de Abandono')
plt.legend(title='Attrition_Flag', labels=['Existing Customer', 'Attrited Customer'])
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
datos1['Marital_Status'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'salmon', 'lightgreen', 'orange', 'lightblue', 'pink'])
plt.title('ProporciÃ³n de Clientes por Estado Civil')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
pd.crosstab(datos1['Income_Category'], datos1['Marital_Status']).plot(kind='bar', stacked=True)
plt.xlabel('CategorÃ­a de Ingresos')
plt.ylabel('Cantidad de Clientes')
plt.title('DistribuciÃ³n de Clientes por CategorÃ­a de Ingresos y Estado Civil')
plt.legend(title='Estado Civil')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Card_Category', hue='Attrition_Flag', data=datos1)
plt.xlabel('CategorÃ­a de Tarjeta de CrÃ©dito')
plt.ylabel('Frecuencia de Compras')
plt.title('Frecuencia de Compras por CategorÃ­a de Tarjeta de CrÃ©dito')
plt.legend(title='Attrition_Flag', labels=['Existing Customer', 'Attrited Customer'])
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
datos1['Income_Category'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('DistribuciÃ³n de Clientes por CategorÃ­a de Ingresos')
plt.show()
```
