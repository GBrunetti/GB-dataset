## *3 - Descripcion de las variables*


- **CLIENTNUM**: Número de identificación único asignado a cada cliente en la base de datos. Este número es utilizado para identificar a cada cliente de manera exclusiva y no tiene relación directa con las características o comportamientos del cliente.

- **Attrition_Flag**: Indicador de si el cliente ha abandonado (Attrited Customer) o sigue siendo cliente activo (Existing Customer). Esta variable es categórica y es crucial para analizar y predecir la atrición de clientes. Ésta será, por tanto, nuestra variable objetivo en el modelo de predicción.

- **Customer_Age**: Edad del cliente en años. Esta variable numérica puede proporcionar información sobre la demografía de los clientes y su relación con el comportamiento de atrición.

- **Gender**: Género del cliente, categorizado como 'M' para masculino y 'F' para femenino. Esta variable categórica puede ayudar a segmentar el análisis demográfico y comportamental.

- **Dependent_count**: Número de personas dependientes del cliente. Esta variable numérica puede indicar el nivel de responsabilidad financiera del cliente.

- **Education_Level**: Nivel de educación alcanzado por el cliente, categorizado en diferentes niveles como 'Uneducated', 'High School', 'College', 'Graduate', 'Post-Graduate', 'Doctorate' y 'Unknown'. Esta variable categórica puede influir en el análisis del perfil de clientes.

- **Marital_Status**: Estado civil del cliente, categorizado como 'Single', 'Married', 'Divorced', 'Unknown'. Esta variable puede ayudar a segmentar a los clientes según su estado civil.

- Income_Category: Categoría de ingresos del cliente, clasificada en rangos como Less than \$40K, \$40K - \$60K, \$60K - \$80K, \$80K - \$120K, \$120K +, y 'Unknown'. Esta variable categórica es importante para analizar el poder adquisitivo y comportamiento financiero de los clientes

- **Card_Category**: Categoría de la tarjeta de crédito del cliente, como 'Blue', 'Silver', 'Gold', 'Platinum'. Esta variable categórica puede influir en la relación del cliente con el banco y su comportamiento de gasto.

- **Months_on_book**: Número de meses que el cliente ha estado registrado en el banco. Esta variable numérica indica la antigüedad del cliente y puede ser crucial para el análisis de la lealtad y atrición.

- **Total_Relationship_Count**: Total de productos financieros que el cliente tiene con el banco, como cuentas corrientes, de ahorro, préstamos, etc. Esta variable numérica puede indicar el nivel de relación del cliente con el banco.

- **Months_Inactive_12_mon**: Número de meses en los últimos 12 meses en los que el cliente ha estado inactivo. Esta variable numérica puede ayudar a identificar patrones de inactividad y riesgo de atrición.

- **Contacts_Count_12_mon**: Número de contactos que el banco ha tenido con el cliente en los últimos 12 meses. Esta variable numérica puede indicar el nivel de interacción y compromiso del cliente.

- **Credit_Limit**: Límite de crédito total asignado al cliente. Esta variable numérica es importante para analizar el perfil de crédito del cliente y su capacidad de gasto.

- **Total_Revolving_Bal**: Saldo revolvente total del cliente en la tarjeta de crédito. Esta variable numérica puede indicar el uso de crédito del cliente y su comportamiento financiero.

- **Avg_Open_To_Buy**: Promedio de crédito disponible para el cliente, calculado como la diferencia entre el límite de crédito y el saldo revolvente. Esta variable numérica puede indicar la capacidad de gasto adicional del cliente.

- **Total_Amt_Chng_Q4_Q1**: Cambio en el monto total de transacciones realizadas por el cliente entre el cuarto trimestre y el primer trimestre. Esta variable numérica, expresada como una fracción, puede indicar cambios en el comportamiento de gasto del cliente.

- **Total_Trans_Amt**: Monto total de transacciones realizadas por el cliente. Esta variable numérica puede indicar el nivel de actividad financiera del cliente.

- **Total_Trans_Ct**: Número total de transacciones realizadas por el cliente. Esta variable numérica puede indicar la frecuencia de uso de la tarjeta de crédito.

- **Total_Ct_Chng_Q4_Q1**: Cambio en el número total de transacciones realizadas por el cliente entre el cuarto trimestre y el primer trimestre. Esta variable numérica, expresada como una fracción, puede indicar cambios en el comportamiento de uso de la tarjeta de crédito.

- **Avg_Utilization_Ratio**: Promedio del ratio de utilización del crédito del cliente, calculado como el saldo revolvente dividido por el límite de crédito. Esta variable numérica puede indicar qué tan cerca está el cliente de utilizar su límite de crédito.

- **Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1**: Variable utilizada para el cálculo interno del modelo de Machine Learning.

# **D -	ELECCION DE ALGORITMO DE REGRESION O CLASIFICACION**

Optamos por enfocarnos en algoritmos de clasificación, ya que la variable objetivo (Attrition_Flag) es binaria. A continuación se detallan los pasos y la justificación de la elección del algoritmo, junto con la implementación y evaluación del modelo.

Dado que nuestro objetivo es predecir si un cliente abandonará los servicios del banco (sí/no), los algoritmos de clasificación son más adecuados. Consideraremos 3 algoritmos de clasificación para en una proxima etapa evaluar su desempeño a fin de seleccionar el mejor modelo:

## *1 - Regresión Logística*

### I - Preparación de Datos

Las variables categóricas se convertiran a numéricas usando OneHotEncoding.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

Seleccionar características y objetivo

```python
X = datos.drop(columns=['Attrition_Flag', 'Attrition_Flag_Bin'])
y = datos['Attrition_Flag_Bin']
```

Identificar columnas categóricas y numéricas

```python
categorical_cols = X.select_dtypes(include=['object']).columns
numerical_cols = X.select_dtypes(include=['float64', 'int64']).columns
```

Crear el transformador para las columnas categóricas

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', numerical_cols),
        ('cat', OneHotEncoder(), categorical_cols)
    ])
```

Dividir el conjunto de datos en entrenamiento y prueba

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### II - Modelo Regresion Logística

Crear el pipeline de preprocesamiento y el modelo de regresión logística

```python
log_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000))
])
```

Entrenar el modelo de Regresión Logística

```python
log_pipeline.fit(X_train, y_train)
```

## *2 - Bosques Aleatorios (Random Forest)*

```python
from sklearn.ensemble import RandomForestClassifier

```

###Modelo de Bosques Aleatorios

Crear el pipeline de preprocesamiento y el modelo de Bosques Aleatorios

```python
rf_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])
```

Entrenar el modelo

```python
rf_pipeline.fit(X_train, y_train)
```

## *3 - Gradient Boosting (XGBoost)*


```python
from xgboost import XGBClassifier
```

###Modelo Gradient Boosting

Crear el pipeline de preprocesamiento y el modelo de XGBoost

```python
xgb_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42))
])
```

Entrenar el modelo

```python
xgb_pipeline.fit(X_train, y_train)
```

# **E -	CALCULOS Y ANÁLISIS DE METRICAS BASICAS PARA VALIDAR EL MODELO**

Para validar los modelos de clasificación entrenados, se utilizan varias métricas de evaluación que ayudan a medir su rendimiento. Las métricas básicas que se calcularán para los modelos son las siguientes:

**I - Accuracy (Exactitud):** Proporción de predicciones correctas sobre el total de predicciones.

**II - Precision (Precisión):** Proporción de verdaderos positivos sobre el total de positivos predichos.

**III - Recall (Sensibilidad):** Proporción de verdaderos positivos sobre el total de verdaderos positivos reales.

**IV - F1-Score:** Media armónica de precisión y recall

**V - AUC-ROC:** Área bajo la curva ROC, que evalúa la capacidad del modelo para distinguir entre las clases




## *1 - Metricas y resultados del Modelo de Regresion Logistica*

### Predicciones y evaluación del modelo

```python
y_pred_log = log_pipeline.predict(X_test)
print("Regresión Logística:")
print(classification_report(y_test, y_pred_log))
print("Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred_log))
print("AUC-ROC:", roc_auc_score(y_test, y_pred_log))
```

## *2 - Metricas y resultados del Modelo de Bosques Aleatorios*

### Predicciones y evaluación del modelo

```python
y_pred_rf = rf_pipeline.predict(X_test)
print("Bosques Aleatorios:")
print(classification_report(y_test, y_pred_rf))
print("Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred_rf))
print("AUC-ROC:", roc_auc_score(y_test, y_pred_rf))
```

## *3 - Metricas y resultados del Modelo Gradient Boost*

### Predicciones y evaluación del modelo

```python
y_pred_xgb = xgb_pipeline.predict(X_test)
print("XGBoost:")
print(classification_report(y_test, y_pred_xgb))
print("Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred_xgb))
print("AUC-ROC:", roc_auc_score(y_test, y_pred_xgb))
```
